```python
import random
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
import plotly.express as px

x = [0]
y = [0]


current_x = 0
current_y = 0

for i in range(100000):
    step = random.randint(0,1)*2-1
    if random.random() > .5:
        current_x += step
    else:
        current_y += step

    x.append(current_x)
    y.append(current_y)

plt.plot(x,y)
plt.show()
```
